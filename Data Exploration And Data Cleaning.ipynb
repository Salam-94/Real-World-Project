## Imputing Null Values With Central Tendency for "totalyearlycompensation" and "basesalary"

```python
sns.boxplot(df.basesalary) #Checking the Distributio of Data with boxplot for "basesalary" column
```

```python
sns.boxplot(df.totalyearlycompensation) #Checking the Distributio of Data with boxplot for "totalyearlycompensation" column
```

```python
list1= [df.basesalary ,df.totalyearlycompensation]  #Againg checking the Distribution for both column by using Loop
 
for i in list1:
    sns.boxplot(i)
    plt.show() 
```

```python
sns.distplot(df.basesalary) #checking the distribution 
```

```python
sns.distplot(df.totalyearlycompensation)
```

```python
df['basesalary'] = df['basesalary'].fillna(df['basesalary'].median())  #imputing null values using median for both columns one by one
df['totalyearlycompensation'] = df['totalyearlycompensation'].fillna(df['totalyearlycompensation'].median())
```
